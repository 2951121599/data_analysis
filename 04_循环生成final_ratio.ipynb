```python
import numpy as np
import pandas as pd
```

```python
for i in range(len(pd.read_csv('final_is_area.csv').values)):  # 循环几个内标
    print(i)
    input_data1 = pd.read_csv('final_area.csv').values  # 读文件 不包含header
    input_data2 = pd.read_csv('final_is_area.csv').values  # 读文件 不包含header
    col = pd.read_csv('final_area.csv',header=None).values[1:,0:2]  # 前两列数据(去除第一行)
    row = pd.read_csv('final_area.csv',header=None).values[0,:]  # 第一行行名
    feature = input_data2[i,0:1]  # 获取IS内标名称 组成生成的文件名
    feature = feature[0]
    print("feature----------", feature)
    input_data1 = input_data1[0:,2:]  # 纯数据
    input_data2 = input_data2[i,2:]  # 纯数据(一行IS内标值)
    print("input_data1----------", input_data1)
    print("input_data2----------", input_data2)

    res = np.divide(input_data1,input_data2)  # 做矩阵除法
    res = np.column_stack((col,res))  # 合并列
    res = np.row_stack((row,res))  # 合并行
    print("res----------", res)
    filename = 'final_is_ratio_' + feature +  '.csv'  # 组成生成的文件名
    np.savetxt(filename, res, fmt='%s', delimiter=',')  # numpy.array写文件
```
